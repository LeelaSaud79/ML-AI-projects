```python
import requests
```

```python
data = requests.get("https://www.tutorialsfreak.com/")
data
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(data.content, "html.parser")
print(soup.prettify())
```

```python
soup.find_all("p")
```

```python
soup.find_all("h2")
```

```python
class_Data = soup.find("div", class_ = "bg-white border-0 p-3 card-footer")
class_Data
```

```python
class_Data.find_all("p")
```

How to get texts?

```python
lines = soup.find_all("p")
lines
```

```python
for l in lines:
    print(l.text)
```
